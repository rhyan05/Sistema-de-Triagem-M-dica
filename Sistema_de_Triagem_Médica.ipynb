```python
def avaliar_triagem(sintomas, temperatura, pressao_arterial, frequencia_cardiaca):
    sintomas_urgentes = {"Dor no peito", "Falta de ar", "Desmaio", "Confusão mental"}
    sintomas_moderados = {"Dor abdominal", "Sangramento", "Febre alta", "Dor de cabeça intensa"}

    sintomas_agrupados = set(sintomas)

    if sintomas_agrupados & sintomas_urgentes:
        return "Emergência! Procure atendimento médico imediatamente."

    if temperatura >= 39.0 or frequencia_cardiaca > 130 or pressao_arterial < 100:
        return "Urgente - Procure atendimento médico imediatamente."

    if sintomas_agrupados & sintomas_moderados:
        return "Moderado - Precisa de médico, mas não é urgente."

    return "Leve - Não é urgente, mas deve ser avaliado por um médico."


print("Sistema de Triagem de Emergência")

sintomas_disponiveis = [
    "Dor no peito",
    "Falta de ar",
    "Dor abdominal",
    "Sangramento",
    "Desmaio",
    "Confusão mental",
    "Febre alta",
    "Dor de cabeça intensa",
    "Dificuldade para falar",
    "Dor nas costas",
    "Dor nas articulações",
    "Vômito",
    "Diarreia"
]

print("\nSelecione os sintomas, separados por vírgulas (digite os números):")
for i, sintoma in enumerate(sintomas_disponiveis, start=1):
    print(f"{i}. {sintoma}")

escolha_sintomas = input("Digite os números dos sintomas escolhidos: ")

sintomas = []
for escolha in escolha_sintomas.split(","):
    try:
        numero = int(escolha.strip())
        if 1 <= numero <= len(sintomas_disponiveis):
            sintomas.append(sintomas_disponiveis[numero - 1])
    except ValueError:
        continue  # ignora entradas inválidas

print("\nAgora, informe os sinais vitais do paciente:")

# Validação da temperatura
while True:
    try:
        temperatura = float(input("Temperatura (°C): "))
        if temperatura < 30 or temperatura > 45:
            print("Temperatura inválida. Insira um valor entre 30°C e 45°C.")
        else:
            break
    except ValueError:
        print("Valor inválido. Digite um número.")

# Validação da pressão arterial
while True:
    try:
        pressao_arterial = int(input("Pressão arterial sistólica (ex: 120): "))
        if pressao_arterial <= 0:
            print("Pressão arterial inválida. Digite um valor positivo.")
        else:
            break
    except ValueError:
        print("Valor inválido. Digite um número inteiro.")

# Validação da frequência cardíaca
while True:
    try:
        frequencia_cardiaca = int(input("Frequência cardíaca (bpm): "))
        if frequencia_cardiaca <= 0:
            print("Frequência inválida. Digite um valor positivo.")
        else:
            break
    except ValueError:
        print("Valor inválido. Digite um número inteiro.")

resultado = avaliar_triagem(sintomas, temperatura, pressao_arterial, frequencia_cardiaca)

# Mostrar resultado
print("\nSinais vitais do paciente:")
print(f"Sintomas: {', '.join(sintomas) if sintomas else 'Nenhum selecionado'}")
print(f"Temperatura: {temperatura}°C")
print(f"Pressão arterial: {pressao_arterial} mmHg")
print(f"Frequência cardíaca: {frequencia_cardiaca} bpm")
print(f"\nClassificação de triagem: {resultado}")
```
